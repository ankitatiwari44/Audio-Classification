Sample_rate:

The sample rate in audio refers to the number of times an audio signal is sampled during the conversion of analog audio into digital information. This rate determines how accurately the digital representation of the original sound is captured.

```python
## dog sound
plt.figure(figsize=(14,5))
data,sample_rate = librosa.load(filename)
librosa.display.waveshow(data,sr=sample_rate)
ipd.Audio(filename)
```
